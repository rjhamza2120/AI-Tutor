```python
import os
from langchain_community.document_loaders import RecursiveUrlLoader
from bs4 import BeautifulSoup as Soup
import requests

def test_access():
    try:
        response = requests.get(
            "https://wander-lush.org/most-beautiful-places-in-pakistan/",
            headers={"User-Agent": os.environ.get("USER_AGENT")},
            timeout=10
        )
        print(f"Status Code: {response.status_code}")
        print(f"Content Length: {len(response.text)}")
        return True
    except Exception as e:
        print(f"Direct access failed: {e}")
        return False

# Test access first
if test_access():
    def simple_extractor(html: str) -> str:
        soup = Soup(html, "html.parser")
        return soup.get_text()
    
    try:
        loader = RecursiveUrlLoader(
            url="https://wander-lush.org/most-beautiful-places-in-pakistan/",
            max_depth=1, 
            timeout=30,
            extractor=simple_extractor,
            prevent_outside=True,
            check_response_status=True,
            use_async=False,
            continue_on_failure=True,
        )
        
        data = loader.load()
        print(f"Loaded {len(data)} documents successfully.")
        
        # Check if we have data before accessing
        if len(data) > 0:
            print(f"First document length: {len(data[0].page_content)}")
            print(f"Source: {data[0].metadata}")
        else:
            print("No documents were loaded. Check if the site is blocking requests.")
            
    except Exception as e:
        print(f"Error loading documents: {type(e).__name__}: {str(e)}")
        import traceback
        traceback.print_exc()
```

```python
import os


from langchain_community.document_loaders import RecursiveUrlLoader
from bs4 import BeautifulSoup as Soup

def simple_extractor(html: str) -> str:
    soup = Soup(html, "html.parser")
    return soup.get_text()

loader = RecursiveUrlLoader(
    url="https://www.sunmarke.com/",
    max_depth=2,
    timeout=30,
    extractor=simple_extractor,
    prevent_outside=True,
    continue_on_failure=True,
)

data = loader.load()
print(f"Loaded {len(data)} documents")
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter_character = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    length_function=len
)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter_recursive = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100,
    length_function=len
)
```

```python
all_splits_character = text_splitter_character.split_documents(data)
```

```python
all_splits_recursive = text_splitter_recursive.split_documents(data)
```

```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
from langchain_qdrant import QdrantVectorStore
vector_store = QdrantVectorStore(
    client=qdrant_client,
    collection_name="WEBSITE DATA",
    embedding=embeddings,
)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Load text file
loader = TextLoader("scraped_data.txt", encoding="utf-8")
documents = loader.load()

# Create text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100,
    length_function=len
)

# Split documents
chunks = text_splitter.split_documents(documents)

print(f"✅ Loaded and split into {len(chunks)} chunks")
```
